```python


def clean_text(text):
    # Remove English characters
    text = re.sub(r'[a-zA-Z]', '', text)
    # Remove digits
    text = re.sub(r'\d', '', text)
     # Remove symbols like '.', '/', ',' etc.

    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251]+', '', text)
    return text
```

```python
df['Comments']=df['Comments'].apply(clean_text)
```

```python
# prompt: python code to map emotions given as df[emotions] in the df change anger to sad , sarcasm to neutral,excitement to happy,fear to sad ,affection to neutral ,abusive to sad ,surprise to neutral

def map_emotions(emotion):
  if emotion == 'Anger':
    return 'Sad'
  elif emotion == 'Sarcasm':
    return 'Neutral'
  elif emotion == 'Excitement':
    return 'Happy'
  elif emotion == 'Humour':
    return 'Happy'
  elif emotion == 'Fear':
    return 'Sad'
  elif emotion == 'Affection/Love':
    return 'Neutral'
  elif emotion == 'Abusive':
    return 'Sad'
  elif emotion == 'Surprise':
    return 'Neutral'
  else:
    return emotion

df['Majority Emotion'] = df['Majority Emotion'].apply(map_emotions)
df.head()
```

```python
#Word Embedding using Word2Vec:

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# Tokenize the comments
tokenized_comments = df['Comments'].apply(word_tokenize)

# Train Word2Vec model
w2v_model = Word2Vec(sentences=tokenized_comments, vector_size=100, window=5, min_count=1, workers=4)
```

```python
#Preparing Data for LSTM:

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

# Tokenize comments
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['Comments'])
sequences = tokenizer.texts_to_sequences(df['Comments'])

# Pad sequences
max_len = max([len(seq) for seq in sequences])
padded_sequences = pad_sequences(sequences, maxlen=max_len)

# Prepare target variable
target = df['Emotion']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(padded_sequences, target, test_size=0.2, random_state=42)
```

```python
#Training LSTM Model:

from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

# Define model architecture
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=max_len))
model.add(LSTM(100))
model.add(Dense(11, activation='softmax'))

# Compile model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))



```

```python

loss, accuracy_lstm1 = model.evaluate(X_test, y_test, verbose=0)
print("Accuracy:", accuracy_lstm1)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Function to create document vectors using Word2Vec embeddings
def document_vector(word2vec_model, tokenized_document):
    # Initialize an empty array for the document vector
    doc_vector = np.zeros(word2vec_model.vector_size)
    # Initialize a counter to keep track of the number of words in the document
    num_words = 0
    # Loop through each word in the tokenized document
    for word in tokenized_document:
        # If the word is in the vocabulary of the Word2Vec model
        if word in word2vec_model.wv.key_to_index:
            # Add the word vector to the document vector
            doc_vector += word2vec_model.wv[word]
            # Increment the counter
            num_words += 1
    # If at least one word in the document is in the vocabulary
    if num_words > 0:
        # Normalize the document vector by dividing by the number of words
        doc_vector /= num_words
    return doc_vector

# Create document vectors for each tokenized comment
X = np.array([document_vector(w2v_model, comment) for comment in tokenized_comments])

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, df['Emotion'], test_size=0.2, random_state=42)

# Initialize SVM classifier
svm_classifier = SVC()

# Train the classifier
svm_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_classifier.predict(X_test)

# Calculate accuracy
accuracy_svm1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_svm1)
```
